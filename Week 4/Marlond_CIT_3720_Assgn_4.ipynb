# Question 3

* Do we have a class imbalance problem? Please evaluate this and determine the best course of action.
* Now that we have our list of most impactful features, split the dataset into a training and testing dataset. We can use 70% of the data for training and 30% of the data for validation.
* Perform a train/test split

## Hint -
* Pandas has ability to sample records, but you will have to group the data firset
```
df.groupby('...').sample(N)
```

````python
"""
Answer - Question 3
"""
from sklearn.model_selection import train_test_split

class_frequencies = df['RainTomorrow_Yes'].value_counts()
# We can see that there is at least 3 times more false than true that rainTomorrow
# There 45507 false and 12729 true
print(class_frequencies)
df = df.groupby('RainTomorrow_Yes').sample(n=12729, random_state=42)

Y = df.iloc[:,-1].values
X = df.drop(["RainTomorrow_Yes"],axis=1)
#splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    #train_size=0.7,
                                                    test_size=0.3,
                                                    random_state=42)
len(Y)
````

# Question 4


* Perform a Logistic Regression
* Since some of our data consists of floats and integer, we might want to scale using a StandardScaler
* You might find that there is convergence problem and you might have to increase the max iterations of the LR model using max_iter=N parameter

## Hint -
* Sklearn allows us to implement a pipe to pipe several operations together
```
pipe = make_pipeline()
```

````python
"""
Answer - Question 4
"""
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipe = make_pipeline(StandardScaler(),LogisticRegression(C=.001, max_iter=295))
pipe.fit(X_train,Y_train)
Y_pred = pipe.predict(X_test)
````
