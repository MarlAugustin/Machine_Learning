## Question 3

* One way to identify frequent users of their phone plan is to determine if their total # of minutes per phone calls exceeds Q3 * IQR of the total number of minuutes of all phone calls -
* Hint -
  * Add up all of the user minutes spent on the phone calls and generate a quartile distribution
  * Create a new column called high_consumption to identify phone plans where the total number of minutes spent on the phone call is > Q3 * IQR. In previous exercise, we found that we can use command ```df['total_minutes'].quantile(0.75) + 1.5 * (df['total_minutes'].quantile(0.75) - df['total_minutes'].quantile(0.25))```
  * New column should be a boolean (1 or 0). 1 when user total_minutes is greater than the above calculation

````python
"""
Answer - Question 3
"""
# The only way that I figuret to add up all the user minutes was by creating a new column
df['total_minutes'] = df['total_day_minutes'] + df['total_night_minutes'] + df['total_eve_minutes'] + df['total_intl_minutes']
total_minutes_IQR = df['total_minutes'].quantile(0.75) - df['total_minutes'].quantile(0.25)
up_minutes = df['total_minutes'].quantile(0.75) + 1.5 * total_minutes_IQR
df['high_consumption'] = df['total_minutes'].gt(up_minutes)
#Once the the high_comsumption columns was made, there was no reason to keep total_minutes
df = df.drop(columns=['total_minutes'])
df.high_consumption=df.high_consumption.astype('category').cat.codes
df
````

# Question 4

Let's identify which time of the day each account spends most of the time on the phone
* Create a new column(s) in the dataset that would help us identify whether the plan spends more time on the phone during the day, eve, night etcx

#### Hint -
* Since we want numeric columns only, create a three column (high_use_day, high_use_eve, high_use_night). This should be one-hot value - so if person talkm most during the night, the output would be [0,0,1] for columns [high_use_day, high_use_eve, high_use_night]


````python
"""
Answer - Question 4
"""

df['high_use_day'] = df.total_day_minutes.gt(df.total_eve_minutes) & df.total_day_minutes.gt(df.total_night_minutes)
df['high_use_eve'] = df.total_eve_minutes.gt(df.total_day_minutes) & df.total_eve_minutes.gt(df.total_night_minutes)
df['high_use_night'] = df.total_night_minutes.gt(df.total_day_minutes) & df.total_night_minutes.gt(df.total_eve_minutes)
#turning the boolean to int
df.high_use_day=df.high_use_day.astype('category').cat.codes
df.high_use_eve=df.high_use_eve.astype('category').cat.codes
df.high_use_night=df.high_use_night.astype('category').cat.codes

df
````
