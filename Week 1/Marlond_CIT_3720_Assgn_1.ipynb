## Question 5
If we do not know if the data is normally distributed, we can use IQR * 1.5 to determine if we have outliers in the data. A commonly used rule says that a data point is an outlier if it is more than 1.5 ⋅ IQR ‍ above the third quartile or below the first quartile. Said differently, low outliers are below Q 1 − 1.5 ⋅ IQR ‍ and high outliers are above Q 3 + 1.5 ⋅ IQR. IQR = Q3 - Q1.

Determine the number of outliers for species "Iris-versicolor" for each of the four measurements.


````python
"""
Answer - Question 5
"""
iris_versicolor = df[df.species =='Iris-versicolor']
#finding the number of low and high outliers for the sepal length of Iris-versicolor
sep_length_q1 = iris_versicolor['sepal_length'].quantile(0.25)
sep_length_q3 = iris_versicolor['sepal_length'].quantile(0.75)
sep_length_iqr = sep_length_q3 - sep_length_q1
sep_length_low_outlier=len(iris_versicolor[iris_versicolor.sepal_length < (sep_length_q1 - 1.5 * sep_length_iqr)])
sep_length_high_outlier=len(iris_versicolor[iris_versicolor.sepal_length > (sep_length_q3 + 1.5 * sep_length_iqr)])
print("Total number of outlier for the sepal length:", (sep_length_low_outlier+ sep_length_high_outlier))

#finding the number of low and high outliers for the sepal width of Iris-versicolor
sep_width_q1 = iris_versicolor['sepal_width'].quantile(0.25)
sep_width_q3 = iris_versicolor['sepal_width'].quantile(0.75)
sep_width_iqr = sep_width_q3 - sep_width_q1
sep_width_low_outlier=len(iris_versicolor[iris_versicolor.sepal_width < (sep_width_q1 - 1.5 * sep_width_iqr)])
sep_width_high_outlier=len(iris_versicolor[iris_versicolor.sepal_width > (sep_width_q3 + 1.5 * sep_width_iqr)])
print("Total number of outlier for the sepal width:", (sep_width_low_outlier + sep_width_high_outlier))

#finding the number of low and high outliers for the petal length of Iris-versicolor
pet_length_q1 = iris_versicolor['petal_length'].quantile(0.25)
pet_length_q3 = iris_versicolor['petal_length'].quantile(0.75)
pet_length_iqr = pet_length_q3 - pet_length_q1
pet_length_low_outlier=len(iris_versicolor[iris_versicolor.petal_length < (pet_length_q1 - 1.5 * pet_length_iqr)])
pet_length_high_outlier=len(iris_versicolor[iris_versicolor.petal_length > (pet_length_q3 + 1.5 * pet_length_iqr)])
print("Total number of outlier for the petal length:", (pet_length_low_outlier+ pet_length_high_outlier))

#finding the number of low and high outliers for the petal width of Iris-versicolor
pet_width_q1 = iris_versicolor['petal_width'].quantile(0.25)
pet_width_q3 = iris_versicolor['petal_width'].quantile(0.75)
pet_width_iqr = pet_width_q3 - pet_width_q1
pet_width_low_outlier=len(iris_versicolor[iris_versicolor.petal_width < (pet_width_q1 - 1.5 * pet_width_iqr)])
pet_width_high_outlier=len(iris_versicolor[iris_versicolor.petal_width > (pet_width_q3 + 1.5 * pet_width_iqr)])
print("Total number of outlier for the petal width:", (pet_width_low_outlier + pet_width_high_outlier))
````
